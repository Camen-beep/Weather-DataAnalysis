```python
from datetime import datetime
import meteostat as mt
import pandas as pd
```

```python
start = datetime(1990, 1, 1)
end = datetime(2025, 8, 25)
```

```python
location_id = '45005'
```

```python
data = mt.Daily(location_id, start, end)
data = data.fetch()
```

```python
df = pd.DataFrame(data)
```

```python
df['real_time'] = df.index
```

```python
df['real_time_year'] = df['real_time'].dt.year
```

```python
df.groupby('real_time')
```

```python
df.groupby('real_time_year').agg({'tavg':'median'})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='real_time_year', y='tavg')  # Adjust columns as needed
plt.title('Temperature Over Time')
plt.xlabel('real_time_year')
plt.ylabel('tavg')
plt.grid(True)
plt.show()
```

```python
def season(input):
    if  3 <= input <= 5:
        return 'spring'
    elif 6 <= input <= 8:
        return 'summer'
    elif 9 <= input <= 11:
        return 'autumn'
    elif input == 12 or 1 or 2:
        return 'winter'

season(3)
```

```python
df['real_time_month'] = df['real_time'].dt.month
df['real_time_month']
```

```python
df['real_time_season'] = df['real_time_month'].apply(season)
df['real_time_season']
```

```python
df.groupby(['real_time_season', 'real_time_year'])
```

```python
df.groupby(['real_time_season', 'real_time_year']).agg({'tavg':'median'})
```
